```python
import tensorflow as tf

interpreter = tf.lite.Interpreter(model_path="models\ASR_STT\coqui_stt\coqui_stt_asr_model.tflite")
interpreter.allocate_tensors()

print(interpreter.get_input_details())
print(interpreter.get_output_details())


for tensor in interpreter.get_tensor_details():
    print(tensor['name'], tensor['dtype'])
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model(r"models/ASR_STT/coqui_stt")

converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]

tflite_quant_model = converter.convert()

open("models\ASR_STT\coqui_stt\coqui_int8.tflite","wb").write(tflite_quant_model)
```
